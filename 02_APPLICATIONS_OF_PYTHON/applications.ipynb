```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup

# Function to fetch the raw HTML of the webpage
def fetch_raw_html(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Check for HTTP errors
        return response.text  # Return the raw HTML content
    except requests.exceptions.RequestException as e:
        print(f"Error fetching the URL: {e}")
        return None

# Main function to execute the script
def main():
    url = "https://www.geeksforgeeks.org/web-crawling-tools/"
    raw_html = fetch_raw_html(url)
    if raw_html:
        print(raw_html)  # Print the raw HTML content
        # If you want to parse the HTML with BeautifulSoup:
        soup = BeautifulSoup(raw_html, "html.parser")
        # You can now use 'soup' to extract specific elements as needed
    return raw_html
if __name__ == "__main__":
    raw_html = main()
```

```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup

# Simple web crawler to fetch titles from a webpage
url = "https://www.geeksforgeeks.org/web-crawling-tools/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
titles = soup.find_all("title")
h1 = soup.find_all("h1")
p = soup.find_all("p")

print("Titles:", [title.text for title in titles])
print("h1:", [h.text for h in h1])
print("p:", [_.text for _ in p])
```
