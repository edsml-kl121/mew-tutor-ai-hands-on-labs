Given a sufficiently large number of independent samples from any population with a finite mean and variance, the distribution of the sample means will approximate a normal distribution (Gaussian), regardless of the population’s original distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Settings
sample_size = 30      # points per sample
n_samples = 10000     # number of samples


# Generate distributions
uniform_data = np.random.uniform(low=0, high=10, size=(n_samples, sample_size))
# -------------

exp_data = np.random.exponential(scale=2.0, size=(n_samples, sample_size))
# -
#  -
#    -- -----


# Sample means
uniform_means = np.mean(uniform_data, axis=1)
exp_means = np.mean(exp_data, axis=1)

# Variance of original data and sample means
uniform_variance_original = np.var(uniform_data)
uniform_variance_means = np.var(uniform_means)

exp_variance_original = np.var(exp_data)
exp_variance_means = np.var(exp_means)

# Print variances
print("Uniform Distribution:")
print(f"  Variance of original data: {uniform_variance_original:.4f}")
print(f"  Variance of sample means:  {uniform_variance_means:.4f}\n")

print("Exponential Distribution:")
print(f"  Variance of original data: {exp_variance_original:.4f}")
print(f"  Variance of sample means:  {exp_variance_means:.4f}")

# Plotting
plt.figure(figsize=(12, 5))

# Uniform
plt.subplot(1, 2, 1)
plt.hist(uniform_means, bins=50, color='skyblue', edgecolor='black', density=True)
plt.title("CLT: Sample Means from Uniform Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Density")

# Exponential
plt.subplot(1, 2, 2)
plt.hist(exp_means, bins=50, color='salmon', edgecolor='black', density=True)
plt.title("CLT: Sample Means from Exponential Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Density")

plt.tight_layout()
plt.show()
```

•	The variance of the sample means is much smaller than the variance of the original data.
•	It roughly equals the original variance divided by the sample size (law of large numbers), e.g.,
